```python
# notebooks/get_data.ipynb
# -*- coding: utf-8 -*-

'''This ipynb file has 4 steps.

1. Load all data.
2. Make mne data and print info.
3. Turn data into dataset.
4. Visualize EEG data and events.
'''

import os
import sys
sys.path.append("..")  

from pymatreader import read_mat
from utils.data_io import make_data, load_data, visualize, mne_data

import warnings
warnings.filterwarnings('ignore')
%matplotlib widget


# step 1
print(f'==== Loading all data ====')
folder = os.path.join("..", "data") 
all_data = load_data(folder)
for subject, files in sorted(all_data.items()):
    for mat_file, mat_data in sorted(files.items()):
        print(f"{subject} -> {mat_file}: keys = {list(mat_data.keys())}")


# step 2
print(f'\n\n==== Mne data information ====')
raw_data = read_mat(os.path.join('..', 'data', 'subject0', 'train1.mat'))
raw = mne_data(raw_data)
print(f'{raw}, {raw.info}')


# step 3
print(f'\n\n==== Processing mne data ====')
src_dir = '..'   
dst_dir = 'dataset'
subject_ids = ['subject0', 'subject1', 'subject2', 'subject3', 'subject4']
for subject_id in subject_ids:
    make_data(src_dir=src_dir, dst_dir=dst_dir, subject_id=subject_id)


# step 4
print(f'\n\n==== Visualizing all data ====')
visualize(raw_data, sfreq=500)
```
